```python
from google.colab import files
import os
import shutil
print("=" * 60)
print("UPLOAD H√åNH ·∫¢NH HOA IRIS")
print("=" * 60)
print("1Ô∏è. Iris Setosa")
print("2Ô∏è. Iris Versicolor")
print("3. Iris Virginica")
uploaded = files.upload()
expected_files = [
    "iris_setosa.jpg",
    "iris_versicolor.jpg",
    "iris_virginica.jpg"
]
uploaded_files = list(uploaded.keys())
success_count = 0
for i, target_name in enumerate(expected_files):
    if i >= len(uploaded_files):
        break
    src = uploaded_files[i]
    shutil.move(src, target_name)
    size_kb = os.path.getsize(target_name) / 1024
    print(f"{target_name} ({size_kb:.1f} KB)")
    success_count += 1
print("\n" + "=" * 60)
if success_count == 3:
    print("ƒê√£ th√†nh c√¥ng.")
else:
    print(f"Ch·ªâ upload ƒë∆∞·ª£c {success_count}/3 ·∫£nh")
print("=" * 60)
```
